### Library Imports

```python
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import classification_report 
```

### Notebook Imports

```python
import pickle

# Variable for prediction
with open('pkl/X_pred.pkl', 'rb') as f:
    X_pred = pickle.load(f)

with open('pkl/models.pkl', 'rb') as f:
    t_models = pickle.load(f)

with open('pkl/model_training.pkl', 'rb') as f:
    X_test, y_test, X_train, y_train = pickle.load(f)
```

### Predictions

```python
output = t_models['knn'].predict(X_pred)
print ('Predicción de Género: ',[output[0]])

output = t_models['rforest'].predict(X_pred)
print ('Predicción de Género: ',[output[0]])

output = t_models['svm'].predict(X_pred)
print ('Predicción de Género: ',[output[0]])

output = t_models['lg'].predict(X_pred)
print ('Predicción de Género: ',[output[0]])

print ('\nGeneros Posibles(KNN): ' ,  t_models['knn'].predict_proba(X_pred), '\n')
print ('Generos Posibles(Random Forest): ' ,  t_models['rforest'].predict_proba(X_pred), '\n')
print ('Generos Posibles(SVM): ' ,  t_models['svm'].predict_proba(X_pred), '\n')
print ('Generos Posibles(LG): ' ,  t_models['lg'].predict_proba(X_pred), '\n')
```

### Confussion Matrix of Selected Model: **KNN**

```python
fig, ax = plt.subplots(figsize=(11, 11))
ConfusionMatrixDisplay.from_estimator(t_models['knn'], X_test, y_test,ax=ax,cmap=plt.cm.Blues)

print(classification_report(y_train, t_models['knn'].predict(X_train), digits=3))
```
